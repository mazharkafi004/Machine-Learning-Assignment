# **Importing K Fold, declaring number of folds**

```python
from sklearn.model_selection import StratifiedKFold as kfold
folds = kfold(n_splits=10)
folds
```

**Arrays for storing precision and accuracy for each fold**

```python
precisionScores = []
accuracyScores = []
```

# **Neural Network function for K-fold validation**

```python
def ann_model(x_train,y_train,x_test,y_test,epochs,batch_size) :
    model = Sequential();
    model.add(Dense(8, input_shape=(8,), activation='relu'))
    model.add(Dense(6, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile( optimizer = 'adam' , loss = 'binary_crossentropy', metrics = ['accuracy'])
    model.fit(x_train,y_train,epochs=epochs,batch_size = batch_size)
    yp = model.predict(x_test)
    y_pred = [] ;
    for element in yp:
        if element > 0.5 :
            y_pred.append(1)
        else :
            y_pred.append(0)
    print(classification_report(y_test,y_pred))
    precisionScores.append(precision_score(y_test, y_pred))
    accuracyScores.append(accuracy_score(y_test, y_pred))
 
```

# **Neural Network function without K-Fold validation**

```python
def ann_model_without_k_fold(X,y,epochs,batch_size) :
    from sklearn.model_selection import train_test_split
    X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)
    model = Sequential();
    model.add(Dense(8, input_shape=(8,), activation='relu'))
    model.add(Dense(6, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile( optimizer = 'adam' , loss = 'binary_crossentropy', metrics = ['accuracy'])
    model.fit(X_train,y_train,epochs=epochs,batch_size = batch_size)
    yp = model.predict(X_test);
    y_pred = [] ;
    for element in yp:
        if element > 0.5 :
            y_pred.append(1)
        else :
            y_pred.append(0)
    precisionScore = precision_score(y_test, y_pred)
    accuracyScore = accuracy_score(y_test, y_pred)
    return precisionScore,accuracyScore ;
     
```

```python
from sklearn.metrics import confusion_matrix , classification_report, precision_score, accuracy_score
```

# **Unscaled dataset without K FOLD validation**

```python
precisionScore,accuracyScore = ann_model_without_k_fold(X,y,100,10)
```

```python
print("Precision is (with unsclaed and without K-Fold) :  %.2f%% " % (precisionScore))
print("Accuracy is (with unsclaed and without K-Fold) :  %.2f%% " % (accuracyScore))
```

# **Scaled dataset with K-FOLD Validation**

```python
for train_index, test_index in folds.split(scaled_X,scaled_y):
   X_train, X_test = scaled_X[train_index], scaled_X[test_index]
   y_train, y_test = scaled_y[train_index], scaled_y[test_index]
   ann_model(X_train,y_train,X_test,y_test,100,10)
```
